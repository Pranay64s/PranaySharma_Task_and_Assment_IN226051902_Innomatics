```python
# Problem 1

def analyze_engagement(likes):
    total_likes = 0
    for like in likes:
        total_likes += like

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)

analyze_engagement([200, 300, 250, 300])
```

```python
# Problem 2

def check_stock(stock):
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    print("Medicine Stock:", stock)
    print("Status:", status)


check_stock(6)
```

```python
# Problem 3

def rainfall_checker(rainfall_data, required_level):
    total = 0
    for rain in rainfall_data:
        total += rain

    average = total / len(rainfall_data)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", average)
    print("Rainfall Status:", status)

rainfall_checker([70, 75, 80, 63], 70)
```

```python
# Problem 4

def detect_duplicates(usernames):
    if len(usernames) != len(set(usernames)):
        result = "Yes"
    else:
        result = "No"

    print("Duplicate Accounts Found:", result)

detect_duplicates(["user1", "user2", "user3", "user1"])
```

```python
# Problem 5

def check_eligibility(age):
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Patient Age:", age)
    print("Eligibility Status:", status)

check_eligibility(21)
```

```python
# Problem 6

def premium_crops(prices):
    premium = []

    for price in prices:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)


premium_crops([1500, 2500, 1800, 3200])
```

```python
# Problem 7

def system_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", errors)
    print("System Status:", status)


system_health(7)
```

```python
# Problem 8

def transaction_checker(amount):
    daily_limit = 50000

    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"

    print("Transaction Amount:", amount)
    print("Transaction Status:", status)

transaction_checker(60000)
```

```python
# Problem 9

def attendance_checker(attendance_list):
    total_classes = len(attendance_list)
    attended = 0

    for status in attendance_list:
        if status == 1:
            attended += 1

    percentage = (attended / total_classes) * 100

    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", eligibility)


attendance_checker([1,1,1,0,1])
```

```python
# Problem 10

def calculate_bill(units):
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage


bill_amount, usage_status = calculate_bill(250)
print("Total Bill:", bill_amount)
print("Usage Classification:", usage_status)
```
